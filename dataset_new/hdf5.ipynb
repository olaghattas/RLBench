```python
import h5py

def print_structure(name, obj):
    """
    This function prints the structure and type of each item in the HDF5 file.
    - name: Name of the object (group or dataset).
    - obj: HDF5 object (group or dataset).
    """
    if isinstance(obj, h5py.Group):
        print(f"Group: {name}")
        for key in obj:
            print_structure(f"{name}/{key}", obj[key])  # Recursively print the group structure
    elif isinstance(obj, h5py.Dataset):
        print(f"Dataset: {name} | Shape: {obj.shape} | Dtype: {obj.dtype}")

def print_all_attributes(group):
    # For each object (group or dataset) in the group, print attributes
    for name, obj in group.items():
        print(f"Object: {name}")
        if isinstance(obj, h5py.Group):
            print(f"  Group {name} Attributes:")
            for attr_name, attr_value in obj.attrs.items():
                print(f"    {attr_name}: {attr_value}")
            # Recursively print attributes of nested groups
            print_all_attributes(obj)
        elif isinstance(obj, h5py.Dataset):
            print(f"  Dataset {name} Attributes:")
            for attr_name, attr_value in obj.attrs.items():
                print(f"    {attr_name}: {attr_value}")
```

```python
# Replace 'your_file.h5' with the path to your HDF5 file
file_path = '/home/olagh48652/segmentation_gib/drawer_lay5_g53b29.hdf5'

with h5py.File(file_path, 'r') as file:
    # print(file["data/demo_0/obs/ee_states"][0])
    for key in file:
        print_structure(key, file[key])
```
